# HAPPY CUSTOMERS

Overview: As we have a Logistics and delivery company which is a startup , aiming to ensure customer satisfaction by understanding the key factors , influencing their happiness. So
identifying customers happiness is a complex challenge, here we have a narrow data for customer feedback through surveys. By utilizing this data , we need to predict customer satisfaction and prioritize operational improvements.
So the survey data captures customer responses (on a scale of 1 to 5) regarding delivery timeliness, order accuracy, completeness, pricing, courier satisfaction, and app usability.The target variable indicates whether a customer is happy (1) or unhappy (0).
The goal is to :
1. Build a predictive model that can accurately classify customer happiness based on their survey responses, achieving at least 73% accuracy.
2. And to identify the most impacting features leading to  customer satisfaction to optimize future surveys and operational strategies.

### Stacking Classifier

```python
from sklearn.ensemble import StackingClassifier
## 1. Create an imbalanced dataset
X, y = make_classification(n_samples=1000, n_classes=2, weights=[0.75, 0.25], 
                          random_state=seed)
## 2. Calculate class weights
from sklearn.utils.class_weight import compute_class_weight
classes = np.unique(y_train)
weights = compute_class_weight('balanced', classes=classes, y=y_train)
class_weights = dict(zip(classes, weights))

# For XGBoost
scale_pos_weight = len(y_train[y_train==0]) / len(y_train[y_train==1])

## 3. Define base models with class weights
base_models = [
    ('lr', make_pipeline(
        StandardScaler(),
        LogisticRegression(class_weight=class_weights, max_iter=1000, random_state=seed)
    )),
    ('svm', make_pipeline(
        StandardScaler(),
        SVC(class_weight=class_weights, kernel='rbf', probability=True, random_state=seed)
    )),
    ('rf', RandomForestClassifier(
        class_weight=class_weights, 
        n_estimators=100, 
        max_depth=5, 
        random_state=seed
    )),
    ('xgb', XGBClassifier(
        scale_pos_weight=scale_pos_weight,
        max_depth=3,
        learning_rate=0.1,
        n_estimators=100,
        random_state=seed
    ))
]

## 4. Define meta-model (final estimator)
meta_model = LogisticRegression(class_weight=class_weights, random_state=seed)

## 5. Create stacking classifier
stacking_model = StackingClassifier(
    estimators=base_models,
    final_estimator=meta_model,
    stack_method='predict_proba',  # Use probabilities for better performance
    cv=5,  # Use 5-fold cross-validation for stacking
    passthrough=False  # Whether to include original features
)

## 6. Train and evaluate
stacking_model.fit(X_train, y_train)
y_pred = stacking_model.predict(X_test)
y_proba = stacking_model.predict_proba(X_test)[:, 1]

print("Stacking Classifier Performance:")
print(classification_report(y_test, y_pred, target_names=['Class 0', 'Class 1']))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

## 7. Compare with base models
print("\nBase Model Comparison:")
for name, model in base_models:
    model.fit(X_train, y_train)
    y_pred_base = model.predict(X_test)
    print(f"\n{name.upper()} Performance:")
    print(classification_report(y_test, y_pred_base, target_names=['Class 0', 'Class 1']))
```
